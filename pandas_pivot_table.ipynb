```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\file_last_access_data\input_data_to_process.txt', low_memory=False)
```

```python
df.head()
```

```python
df['last_access'] = df['last_access'].map(lambda x: x.split('T')[0])  #last access column changed to only date
```

```python
df.head()
```

```python
df.__delitem__('last_modified')  # deleting column as not required the same
```

```python
df.head()
```

```python
df['filepath'] = df['filepath'].map(lambda x: x.split('/')[2])  #file path splitted by '/' and taken out only required item
```

```python
df.head()
```

```python
# print(df[df["Filename"].isnull()])
df[pd.isnull(df).any(axis=1)]   # checking if any Filename value is null
```

```python
df2 = df.dropna()  # deleted the Null values
```

```python
df2.head()
```

```python
processed = df2[df2['Filename'].str.contains('_')]
```

```python
processed.head()
```

```python
non_processed = df2[~df2['Filename'].str.contains('_')]
```

```python
non_processed.head()
```

```python
if df2.shape[0] == processed.shape[0] + non_processed.shape[0]:
    processed_summary = processed.pivot_table(index = ['last_access', 'filepath'], aggfunc='sum')
    non_processed_summary = non_processed.pivot_table(index = ['last_access', 'filepath'], aggfunc='sum')
else:
    print('Looks like some lines missing in processed or non processed files. pls check')

#checking if any lines missed while the data frames splitted. if we did not miss any data, do summary using pivot table.
```

```python
non_processed_summary.head()
```

```python
non_processed_summary.head()
```

```python
if round((processed_summary['Size_GB'].sum() + non_processed_summary['Size_GB'].sum()),2) == round(df2['Size_GB'].sum(),2):
    processed_summary.to_csv(r'C:\file_last_access_data\processed_summary.csv')
    non_processed_summary.to_csv(r'C:\file_last_access_data\non_processed_summary.csv')
else:
    print("Looks like summary does not match with original file. pls check")
    
# checking if the sum of Size_GB in 2 summaries are matching with original file.
# if it matches the output data is correct and saves as csv to location.
```
