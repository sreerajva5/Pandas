```python
import pandas as pd
import numpy as np
```

```python
ser1 = pd.Series([1,2,3,4])
```

```python
df = pd.DataFrame({'col1':[1,2,3,4], 'col2':[20,21,22,23]})
```

```python
print(ser1)
```

```python
# map with series

ser1.map(lambda x:x+2)
```

```python
# map will not work with dataframe

df.map(lambda x: x+2)
```

```python
# apply with series

ser1.apply(lambda x:x+2)
```

```python
# apply with dataframe

df.apply(lambda x:x+2)
```

```python
# applymap with series (will not work)
ser1.applymap(lambda x:x+2)
```

```python
# applymap with dataframe

df.applymap(lambda x : x+2)
```

              map     apply     applymap
series       yes      yes         no
dataframe    no       yes         yes
